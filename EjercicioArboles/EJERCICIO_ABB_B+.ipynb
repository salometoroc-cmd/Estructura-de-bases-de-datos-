```python
import random  # importa ramdom para crear de forma aleatoria

def generar_estudiante(matricula):      # se genera una funcion para generar los estudiantes
    nombres = ["Juan", "Maria","Salome" ,"Susana","Pedro", "Ana", "Luis", "Laura", "Carlos", "Sofia", "Diego", "Valeria"] # nombres con los que se van a generar los estudidantes aleatorios
    apellidos = ["Gomez", "Rodriguez", "Fernandez", "Lopez", "Martinez", "Perez", "Garcia", "Diaz","Hoyos","Agudelo" ,"Sanchez", "Romero"] # apellidos base para generar los estudiantes aleatorios
    nombre = f"{random.choice(nombres)} {random.choice(apellidos)}"   #el nombre se compone de un apellido al azar y uno nombre al azar
    edad = random.randint(18, 28)   # la edad se escoge de forma aleatoria con un rango de una persona universitaria
    promedio = round(random.uniform(70.0, 100.0), 2)   #el promedio  es una tonota que se redondea con solo dos decimales
    return {"matricula": matricula, "nombre": nombre, "edad": edad, "promedio": promedio} #devuelve todo como una libreria

# Generar 10,000 estudiantes
estudiantes = []   # creamos estudiantes como una lista vacia
num_estudiantes = 10000    # vamos a crear 10.000 estudiantes de forma aleatoria
for i in range(1, num_estudiantes + 1): # genera un ciclo for para llamar a la funcion con 10.000 repeitciones para crear a los 10.000 estudiantes
    estudiantes.append(generar_estudiante(i))

print(f"Se han generado {len(estudiantes)} estudiantes.")  #imprimimos que se hizo los estudiantes
# Mostrar los primeros 5 estudiantes para verificar
print("Primeros 5 estudiantes:")   # como ejemplo se muestran los primeros 5 estudiantes para verificar
for i in range(5):
    print(estudiantes[i])
```

```python
def buscar_estudiante_por_id(matricula_buscar, lista_estudiantes):    #vamos a buscar a los estudiantes
    print(f"\nBuscando estudiante con matrícula: {matricula_buscar}")
    for estudiante in lista_estudiantes:   # para cada estudiante en la lista de estudiantes
        if estudiante['matricula'] == matricula_buscar:    #si el estudiante si es devuelve a esa persona
            return estudiante    # es una busqueda lineal
    return None
```

```python
def insertar_estudiante(nuevo_estudiante, lista_estudiantes):
    # Verificar si ya existe un estudiante con la misma matrícula
    for estudiante in lista_estudiantes:    # para cada estudiante en la lista de estudiantes
        if estudiante['matricula'] == nuevo_estudiante['matricula']: # si el estudiante nuevo que se quiere ingresar ya existe pues delvuele falso
            print(f"\nError: Ya existe un estudiante con la matrícula {nuevo_estudiante['matricula']}. No se puede insertar.")
            return False

    # Si no existe, se inserta el nuevo estudiante
    lista_estudiantes.append(nuevo_estudiante)   # se añade ese estudiantes
    print(f"\nEstudiante con matrícula {nuevo_estudiante['matricula']} insertado correctamente.")
    return True
```

```python
class Node:            #se crea una clase
    def __init__(self, key):
        self.key = key # valor que guarda el nodo (la clave o ID)
        self.left = None   # referencia al hijo izquierdo
        self.right = None   # referencia al hijo derecho

class BST:    #clase que representa el arbol binario de busqueda
    def __init__(self):   #contructor
        self.root = None    #la raiz del arbol que está vacia

    def insert(self, key):       #funcion de insertar un valor en el arbol
        self.root = self._insert_recursive(self.root, key)  # llamamos la funcion recuersiva

    def _insert_recursive(self, node, key):   #funcion para insertar un valor en el arbol
        if node is None:       # si el el espacio está vacio  se creea un nuevo nodo
            return Node(key)
        if key < node.key:  # si la clave de nodo es menor , se va al subarbol izquierdo
            node.left = self._insert_recursive(node.left, key)
        elif key > node.key:    # si la clave de nodo es mayor , se va al subarbol derecho
            node.right = self._insert_recursive(node.right, key)
        return node

    def buscar(self, key):   #funcion para llamar a la forma recursiva
        return self._buscar_recursive(self.root, key)

    def _buscar_recursive(self, node, key):
        if node is None or node.key == key:      # Si llegamos a None o encontramos el nodo, terminamos
            return node  # Retorna el nodo si se encuentra, o None si no
        if key < node.key:           # Si la clave es menor, seguimos buscando en el subárbol izquierdo
            return self._buscar_recursive(node.left, key)
        else:   #sino , osea es mayor seguimos en el subarbol derecho
            return self._buscar_recursive(node.right, key)

    def altura(self, nodo): # calculamos la altura para usarla luego
        if nodo is None:   #si el nodo no existe , pues la altura es cero
            return 0
        return 1 + max(self.altura(nodo.left), self.altura(nodo.right))
""" si no la altura es 1 sumado el maximo entre las alturas de izquiera y derecha """

"""recorremos el arbol en orden es decir izquiera luego la raiz y luego el lado de la derecha """
    def inorder_traversal(self):
        elements = []         # lista d eelementos vacia donde se van a guardar los elementos
        self._inorder_recursive(self.root, elements)  #llama a la funcion recursiva
        return elements  #devuelev la lista final

    def _inorder_recursive(self, node, elements):   # está funcion recorre el arbol
        if node: # si el nodo existe
            self._inorder_recursive(node.left, elements)   # se recorre los nodos del lados izquierdo con los valores menores
            elements.append(node.key)
            self._inorder_recursive(node.right, elements)# y luego el lado derecho lo recorre con los valores del lado derecho

print("Clases Node y BST definidas.")
```

#### **Demostración de Árbol B+ con IDs Aleatorios y Ordenados**

Ahora usaremos la implementación más completa del Árbol B+ con los IDs de los estudiantes. Recuerda que, incluso esta versión, es didáctica y omite complejidades de un B+ de producción (como la gestión robusta de la eliminación, la fusión de nodos y algunos aspectos avanzados de rebalanceo y eficiencia de disco).

```python
class BPlusTreeNode:
    def __init__(self, es_hoja=False, orden=3):
        self.es_hoja = es_hoja
        self.claves = []  # Lista de claves
        self.hijos = []   # Lista de hijos (punteros a otros nodos) - Solo para nodos internos
        self.valores = [] # Lista de valores (punteros a datos) - Solo para nodos hoja
        self.siguiente_hoja = None # Puntero al siguiente nodo hoja para búsquedas de rango
        self.padre = None # Puntero al nodo padre
        self.orden = orden # Orden del árbol (número máximo de claves es orden-1)


    def esta_lleno(self):
        return len(self.claves) == self.orden - 1   # Un nodo está lleno cuando tiene el máximo de claves permitido (orden - 1)

    def esta_vacio(self):
        return len(self.claves) < self.orden // 2 ## Un nodo está vacío cuando tiene menos de la mitad de claves

    def __repr__(self):
        return f"[Claves: {self.claves}, Hijos: {[h.claves for h in self.hijos]}, EsHoja: {self.es_hoja}]"  # define cómo se representa un nodo del árbol B+ cuando se imprime o se convierte a string.

class BPlusTree:
    def __init__(self, orden=3):
        self.orden = orden
        self.raiz = BPlusTreeNode(es_hoja=True, orden=orden) # La raíz inicial es una hoja

    def _buscar_hoja(self, clave):
        nodo_actual = self.raiz   # el nodo actual es la raiz
        while not nodo_actual.es_hoja:  # mmientras no lleguemos a una hoja
            i = 0
            #Avanzamos mientras la clave sea mayor o igual
            while i < len(nodo_actual.claves) and clave >= nodo_actual.claves[i]:
                i += 1
            nodo_actual = nodo_actual.hijos[i] # se baja a ese hijo corrspodiente
        return nodo_actual   # se devuelve el nodo encontrado o la hoja encontrada

    def _dividir_nodo_hoja(self, nodo_hoja):
        # Dividir el nodo hoja en dos
        punto_medio = self.orden // 2   #encontramos el punto medio
        nuevo_nodo_hoja = BPlusTreeNode(es_hoja=True, orden=self.orden)  #creamos un nuevo nodoque sera la mitad derecha

        nuevo_nodo_hoja.claves = nodo_hoja.claves[punto_medio:] #Repartir las claves la mitad a la cerecha
        nuevo_nodo_hoja.valores = nodo_hoja.valores[punto_medio:] #Repartir los valores correspondientes
        nodo_hoja.claves = nodo_hoja.claves[:punto_medio] #El nodo original se queda con la mitad izquierda
        nodo_hoja.valores = nodo_hoja.valores[:punto_medio]  #se le asignan los hijos al nuevo nodo

        # Enlazar nodos hoja como una lista
        nuevo_nodo_hoja.siguiente_hoja = nodo_hoja.siguiente_hoja
        nodo_hoja.siguiente_hoja = nuevo_nodo_hoja

        # Promover la clave media al padre (en B+ es la clave más pequeña del nuevo nodo)
        self._insertar_en_padre(nodo_hoja, nuevo_nodo_hoja.claves[0], nuevo_nodo_hoja)

    def _dividir_nodo_interno(self, nodo_interno):
        punto_medio = self.orden // 2
        clave_promovida = nodo_interno.claves[punto_medio] # se selecciona la clave del medio sube

        nuevo_nodo_interno = BPlusTreeNode(es_hoja=False, orden=self.orden) # se crea el nodo interno que será el derecho
        nuevo_nodo_interno.claves = nodo_interno.claves[punto_medio + 1:] # se toman las cvalves despues del putno medio
        nuevo_nodo_interno.hijos = nodo_interno.hijos[punto_medio + 1:]  #se reasignan los hijos distribuyendose

        # Actualizar padres de los hijos del nuevo nodo interno
        for hijo in nuevo_nodo_interno.hijos:
            hijo.padre = nuevo_nodo_interno

        nodo_interno.claves = nodo_interno.claves[:punto_medio]  #actualizan las claves del nodo original osea el izquierdo que quedó como el original y este se queda solo con las claves hasta el punto medio
        nodo_interno.hijos = nodo_interno.hijos[:punto_medio + 1] # se actualizan los hijos del nodo izquierdo o original

        self._insertar_en_padre(nodo_interno, clave_promovida, nuevo_nodo_interno) # se inserta la clave promovida en el padre, Se pasa el nodo izquierdo ya actualizado, la clave que sube, y el nuevo nodo derecho

    def _insertar_en_padre(self, nodo_izquierdo, clave_promovida, nodo_derecho):
        if nodo_izquierdo == self.raiz and nodo_derecho == self.raiz: # Caso inicial, un solo nodo hoja es la raiz
            return # Ya se manejó al crear la raiz

        if nodo_izquierdo.padre is None: # Si el nodo izquierdo no tiene padre, significa que la raíz se dividió
            nueva_raiz = BPlusTreeNode(es_hoja=False, orden=self.orden) #Se crea un nuevo nodo que será la nueva raíz del árbol
            nueva_raiz.claves.append(clave_promovida)  # La clave que subió de la división se convierte en la primera (y única) clave de la nueva raíz
            nueva_raiz.hijos.append(nodo_izquierdo) # el nodo izquiero se convierte en el hijo izquiero de la nueva raiz
            nueva_raiz.hijos.append(nodo_derecho) # el derecho también es hijo
            self.raiz = nueva_raiz
            nodo_izquierdo.padre = nueva_raiz
            nodo_derecho.padre = nueva_raiz
            return

        padre = nodo_izquierdo.padre  # Obtenemos el padre del nodo izquierdo

        # Encontrar la posición para insertar la clave promovida y el nuevo hijo
        i = 0
        while i < len(padre.claves) and clave_promovida > padre.claves[i]:
            i += 1

        padre.claves.insert(i, clave_promovida)
        padre.hijos.insert(i + 1, nodo_derecho) # El nuevo hijo va a la derecha de la clave promovida
        nodo_derecho.padre = padre

        # Si el padre se llena, dividirlo
        if padre.esta_lleno():
            self._dividir_nodo_interno(padre)

    def insertar(self, clave, valor):
        if not self.raiz: # Árbol vacío
            self.raiz = BPlusTreeNode(es_hoja=True, orden=self.orden)  # Como el árbol está vacío, creamos un nodo hoja como raíz
            self.raiz.claves.append(clave) # se inserta la primera clave en la lista de claves
            self.raiz.valores.append(valor) # se inserta el valor en la lsita de valores
            print(f"Insertado (nueva raíz): Clave {clave}")
            return

        nodo_hoja = self._buscar_hoja(clave) # Llamamos a la función auxiliar que encuentra la hoja correcta

        # Insertar en la posición correcta en el nodo hoja
        i = 0
        while i < len(nodo_hoja.claves) and clave > nodo_hoja.claves[i]:   # Mientras no lleguemos al final Y la clave a insertar sea mayor que la clave actual
            i += 1
        nodo_hoja.claves.insert(i, clave) # se coloca la clave en la posicion i y desplazamos las claves a la derecha , las existentes
        nodo_hoja.valores.insert(i, valor) # se insertan los valores en la misma posición
        print(f"Insertado: Clave {clave}")

        if nodo_hoja.esta_lleno():
            self._dividir_nodo_hoja(nodo_hoja)   # si el nodo hoja está lleno se divide llamando a la funcion

    def buscar(self, clave):
        nodo_actual = self._buscar_hoja(clave)
        try:
            indice = nodo_actual.claves.index(clave)
            return nodo_actual.valores[indice] #  El método index() busca la clave en la lista de claves,Si la encuentra, devuelve su posición
        except ValueError:
            return None # Clave no encontrada en este nodo hoja

    def _imprimir_arbol_recursivo(self, nodo, nivel=0):   # se imprime el arbol recurisvo
        s = "  " * nivel + str(nodo.claves)
        if nodo.es_hoja:   # si es nodo hoja  se añade la informacón de valores
            s += f" (Hoja, Valores: {nodo.valores})"
        s += "\n"  #salto de linea

        for hijo in nodo.hijos:  #para cada hijose llama a la funcion unmentando el nivel y se concatena el resultado
            s += self._imprimir_arbol_recursivo(hijo, nivel + 1)
        return s

    def imprimir_arbol(self):
        if not self.raiz:   # verifica si el arbol está vacio, Si no hay raíz, el árbol está vacío
            return "Árbol B+ vacío."
        print("--- Visualización del Árbol B+ ---")
        print(self._imprimir_arbol_recursivo(self.raiz))
```
