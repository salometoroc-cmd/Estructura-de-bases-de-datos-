```python
import hashlib

hash1 = hashlib.sha256(b"hola").hexdigest()
hash2 = hashlib.sha256(b"Hola").hexdigest()
hash3 = hashlib.sha256(b"chao").hexdigest()
hash4 = hashlib.sha256(b"Chao").hexdigest()

concatenar = hash1 + hash2
hash5 = hashlib.sha256(concatenar.encode()).hexdigest()
print(hash5)

concatenar1 = hash3 + hash4
hash6 = hashlib.sha256(concatenar1.encode()).hexdigest()
print(hash6)

final = hash5 + hash6
hash7 =  hashlib.sha256(final.encode()).hexdigest()
print(hash7)
```

```python
import hashlib
def sha256(data):
    return hashlib.sha256(data.encode()).hexdigest()

hash1 = sha256("hola")
hash2 = sha256("Hola")
hash3 = sha256("chao")
hash4 = sha256("Chao")

hash_holas = sha256(hash1 + hash2)
hash_chao = sha256(hash3 + hash4)

hash_final = sha256(hash_holas + hash_chao)
print(hash_final)
```

```python
import hashlib

def sha256(x):
    return hashlib.sha256(x.encode()).hexdigest()

hash1 = sha256("hola")
hash2 = sha256("Hola")    # hojas
hash3 = sha256("chao")
hash4 = sha256("Chao")

nodos1 = sha256(hash1 + hash2)   #nodos
nodos2 = sha256(hash3 + hash4)

raiz = sha256(nodos1 + nodos2)    # raiz

print("Hash raíz:", raiz)
```
