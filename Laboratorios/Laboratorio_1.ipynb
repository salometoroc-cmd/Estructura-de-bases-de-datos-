Problema 2

Dado el Hash root del arbol de mertel determinar el orden de las transacciones que se generan.

reglas las transacciones son conocidas y el numero de transacciones





Cocreado con Gen IA

```python
import hashlib
import itertools

def arbol_merckel(transacciones):
    # Nivel 0: Hashes de cada transacción
    nivel = []
    for tx in transacciones:
        nivel.append(hashlib.sha256(tx.encode()).hexdigest())

    # Subir de nivel sin duplicar
    while len(nivel) > 1:
        nuevo_nivel = []
        for i in range(0, len(nivel), 2):
            if i + 1 < len(nivel):
                # Combinar dos
                nuevo_nivel.append(hashlib.sha256((nivel[i] + nivel[i+1]).encode()).hexdigest())
            else:
                # Subir solo (no duplicar)
                nuevo_nivel.append(nivel[i])
        nivel = nuevo_nivel

    return nivel[0]

# ========================================
# GENERAR UN HASH OBJETIVO NUEVO (conoces el orden)
# ========================================
print("=== PASO 1: Crear un hash objetivo desde un orden conocido ===\n")

# Tú eliges el orden que quieres que el programa adivine
orden_real = ['TransaccionA', 'TransaccionB', 'TransaccionC']  # <--- Cambia el orden si quieres

hash_objetivo = arbol_merckel(orden_real)

print(f"Orden secreto (solo tú lo sabes): {orden_real}")
print(f"Hash objetivo generado: {hash_objetivo}")
print("\n" + "="*60)

# ========================================
# BUSCAR EL ORDEN (el programa adivina)
# ========================================
print("\n=== PASO 2: Buscar qué orden genera ese hash ===\n")

transacciones_conocidas = ['TransaccionA', 'TransaccionB', 'TransaccionC']
encontrado = False

for permutacion in itertools.permutations(transacciones_conocidas):
    orden_actual = list(permutacion)
    hash_actual = arbol_merckel(orden_actual)

    if hash_actual == hash_objetivo:
        print(f"¡ENCONTRADO!")
        print(f"El orden que genera el hash es: {orden_actual}")
        print(f"Hash verificado: {hash_actual}")
        encontrado = True
        break

if encontrado:
    print("\n✓ El programa adivinó correctamente el orden de las transacciones.")
else:
    print("✗ Error: No se encontró el orden.")
```
